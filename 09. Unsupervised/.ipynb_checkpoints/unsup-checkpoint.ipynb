Для <a href='https://thumbs.dreamstime.com/x/two-lorikeet-birds-2293918.jpg'> картинки </a> нужно уменьшить число цветов в палитре.

Для этого нужно выделить **кластеры в пространстве RGB**, объекты соответствуют пикселям изображения. После выделения кластеров, все пиксели, отнесенные в один кластер, заполняются одним цветом; этот цвет может быть центроидом соответствующего кластера, медианным цветом по кластеру.

Попробуйте различные алгоритмы кластеризации:

* KMeans
* DBSCAN
* AgglomerativeClustering

Рассмотрите число кластеров K = 2, 3, 10, 20 (в алгоритмах, где есть такой гиперпараметр).

Для различных кластеризаций оцените и сравните потери от уменьшения цветов при помощи метрики 
<a href='http://scikit-image.org/docs/dev/api/skimage.measure.html#skimage.measure.compare_ssim'> SSIM </a>. Какой способ оказался лучшим?

Задание оценивает в 5 баллов.

Еще 1 балл даётся за запуск алгоритма на выбранных вами картинках - на них результат должен получиться красивым.

Вот вам вспомогательные функции для считывания и вывода изображений:

```python
from skimage.metrics import structural_similarity as ssim
```

```python
import matplotlib.pyplot as plt
import cv2 # нужно поставить себе cv2 через conda install opencv

def draw_picture(image, bgr=False):
    b, g, r = cv2.split(image) # по умолчанию cv2 почему-то отдает цвета в порядке BGR вместо RGB
    new_image = cv2.merge([r, g, b])
    plt.figure(figsize=(7, 5))
    plt.axis('off')
    plt.imshow(new_image)
    plt.show()
```

img - это обычный трехмерный массив цветов пикселей, где две первые координаты - это координаты каждого пикселя, и третья обозначает R, G или B.

```python
img = cv2.imread('birds.jpg') # считать картинку
draw_picture(img) # нарисовать картинку
```

```python
print(img.shape) # сам img - это обычный трехмерный массив 275 x 400 x 3
# вам скорее все кстати понадобится сделать reshape, чтобы перейти от 275x400 к одномерному массиву пикселей
```

```python
X = img.reshape(275 * 400, 3)
X.shape
```

```python
img.dtype
```

```python
from sklearn.cluster import KMeans

def apply_kmeans_clustering(n_clusters):
    kmeans = KMeans(n_clusters=n_clusters)
    y_preds = kmeans.fit_predict(X)

    centers = kmeans.cluster_centers_
    labels = kmeans.labels_
    
    pic = centers[labels].reshape(275, 400, 3).astype('u1')
    print("SSIM:", ssim(img, pic, channel_axis=2))
    draw_picture(pic)
```

```python
apply_kmeans_clustering(2)
```

```python
apply_kmeans_clustering(3)
```

```python
apply_kmeans_clustering(10)
```

```python
apply_kmeans_clustering(20)
```

Кластеризация методом KMeans дала очень хороший результат: разбиение на 20 кластеров помогло сохранить цвета попугаев и отделить их от фона (об этом говорит достаточно высокое значение SSIM)

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=3)
y_preds = dbscan.fit_predict(X)

centers = dbscan.components_
labels = dbscan.labels_

pic = centers[labels].reshape(275, 400, 3).astype('u1')
print(ssim(img, pic, channel_axis=2))
draw_picture(pic)
```

Видим, что DBSCAN не подходит для кластеризации цветов изображения, он помог только отделить попугаев от фона (что тоже может быть полезно)

```python
resized = cv2.resize(img, dsize=(80, 55), interpolation=cv2.INTER_CUBIC)
```

```python
draw_picture(resized)
```

```python
X = resized.reshape((80 * 55, 3))
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import NearestCentroid

def apply_agglomerative_clustering(n_clusters):
    agglom = AgglomerativeClustering(n_clusters=n_clusters, linkage='average')
    y_preds = agglom.fit_predict(X)

    clf = NearestCentroid()
    clf.fit(X, y_preds)

    centers = clf.centroids_
    labels = agglom.labels_

    pic = centers[labels].reshape(55, 80, 3).astype('u1')
    print("SSIM:", ssim(resized, pic, channel_axis=2))
    draw_picture(pic)
```

```python
apply_agglomerative_clustering(2)
```

```python
apply_agglomerative_clustering(3)
```

```python
apply_agglomerative_clustering(10)
```

```python
apply_agglomerative_clustering(20)
```

AgglomerativeClustering тоже дает неплохой результат, но требует значительно бОльших затрат по памяти, поэтому пришлось сильно сжать исходную картику

В нашем случае лучше всего себя показала кластеризация методом KMeans, она дала достаточно похожие картинки и позволила значительно уменьшить число цветов в палитре

```python
import matplotlib.pyplot as plt
def plot_image(image_id):
    # функция рисует изображение номер image_id
    plt.title(names[y[image_id]])
    plt.imshow(255 - lfw_people['images'][image_id], cmap=plt.cm.gray_r, interpolation='nearest')
    
def plot_grid(image_ids, nrows, ncols):
    # функция рисует изображения с id равными image_ids
    # в nrows рядов и ncols столбцов
    fig = plt.gcf()
    fig.set_size_inches(17.5, 8)
    for index, example_id in enumerate(image_ids):
        plt.subplot(nrows, ncols, index + 1)
        plot_image(example_id)
        plt.axis('off')

plot_grid(range(10), nrows=2, ncols=5)# первые 10 лиц
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50, n_jobs=-1).fit(X_train, y_train)
```

```python
y_preds = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_preds)
```

Видим, что модель дает значительно большую точность, чем константное предсказание

```python
depths = np.arange(10, 30)
scores = np.array([])

for depth in depths:
    model = RandomForestClassifier(n_estimators=50, n_jobs=-1, max_depth=depth, random_state=42).fit(X_train, y_train)
    y_preds = model.predict(X_test)
    scores = np.append(scores, accuracy_score(y_test, y_preds))
```

```python
plt.figure(figsize=(15, 7))
plt.plot(depths, scores)
plt.title('Зависимость точности от глубины')
plt.xlabel('Глубина деревьев')
plt.ylabel('Значение accuracy')
```

```python
depth = depths[scores.argmax()]
```

```python
n_samples = np.arange(1, 15)
scores = np.array([])

for samples in n_samples:
    model = RandomForestClassifier(n_estimators=50, n_jobs=-1, max_depth=depth, random_state=42, min_samples_leaf=samples).fit(X_train, y_train)
    y_preds = model.predict(X_test)
    scores = np.append(scores, accuracy_score(y_test, y_preds))
```

```python
plt.figure(figsize=(15, 7))
plt.plot(n_samples, scores)
plt.title('Зависимость точности от числа элементов в листе')
plt.xlabel('Значение min_samples_leaf')
plt.ylabel('Значение accuracy')
```

```python
samples = n_samples[scores.argmax()]
```

```python
model = RandomForestClassifier(n_estimators=50, n_jobs=-1, max_depth=depth, random_state=42, min_samples_leaf=samples).fit(X_train, y_train)
y_preds = model.predict(X_test)
accuracy_score(y_test, y_preds)
```
